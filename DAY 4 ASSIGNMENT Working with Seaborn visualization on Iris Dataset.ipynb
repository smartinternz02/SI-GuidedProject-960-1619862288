# Working with Seaborn visualization on Iris Dataset

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import seaborn as sns
```

```python
iris = sns.load_dataset('iris')
```

```python
iris
```

# SCATTER PLOT

```python
sns.scatterplot(x = 'species', y = 'sepal_length', data = iris)
plt.show()
```

```python
sns.scatterplot(x = 'sepal_length',y='sepal_width', data = iris)
plt.show()
```

```python
sns.scatterplot(x = 'petal_width',y='sepal_width', data = iris)
plt.show()
```

# LINE PLOT

```python
sns.lineplot(x = 'species', y='sepal_width', data = iris)
plt.show()
```

```python
sns.lineplot(x = 'sepal_length', y = 'sepal_width', data = iris)
plt.show()
```

```python
sns.lineplot(x = 'petal_length', y = 'petal_width', data = iris)
plt.show()
```

```python
sns.lineplot(x = 'species', y = 'petal_length', data = iris)
plt.show()
```

# DISTRIBUTION PLOT

```python
sns.distplot(iris['sepal_length'])
plt.show()
```

```python
sns.distplot(iris['petal_width'])
plt.show()
```

# BAR PLOT

```python
sns.barplot(x = 'species', y = 'sepal_length', data = iris)
plt.show()
```

```python
sns.barplot(x = 'species', y = 'petal_length', data = iris)
plt.show()
```

```python
sns.barplot(x = 'species', y = 'sepal_width', data = iris)
plt.show()
```

```python
sns.barplot(x = 'species', y = 'petal_width', data = iris)
plt.show()
```

```python
plt.figure(figsize = (10,10))
sns.barplot(x = 'sepal_length', y = 'petal_width', data= iris)
plt.show()
```

```python
plt.figure(figsize = (10,10))
sns.barplot(x = 'sepal_length', y = 'sepal_width', data= iris)
plt.show()
```

```python
plt.figure(figsize = (10,10))
sns.barplot(x = 'petal_length', y = 'petal_width', data= iris)
plt.show()
```

# BOX PLOT

```python
sns.boxplot(x = 'species', y = 'sepal_width',data = iris)
plt.show()
```

```python
sns.boxplot(x = 'species', y = 'sepal_length',data = iris)
plt.show()
```

```python
sns.boxplot(x = 'species', y = 'petal_width',data = iris)
plt.show()
```

```python
sns.boxplot(x = 'species', y = 'petal_length',data = iris)
plt.show()
```

# CATEGORICAL PLOT

```python
sns.catplot(x = 'species', y = 'sepal_width', data = iris)
plt.show()
```

```python
sns.catplot(x = 'species', y = 'petal_width', data = iris)
plt.show()
```

# PAIR PLOT

```python
sns.pairplot(iris)
plt.show()
```

# CORRELATION HEATMAP 

```python
plt.figure(figsize = (10,10))
sns.heatmap(iris.corr(), annot = True) 
plt.show()
```
